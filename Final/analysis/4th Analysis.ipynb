Different weapons were used for the terrorist attacks.
They all fall under various categories of weapons such as explosives, firearms, biological, etc
The count of weapons is calculated and displayed in the chart with weapon types.

```python
# importing the required variables like pandas, numpy and matplotlib
import os
import pandas as pd
from pandas import DataFrame
import datetime
import calendar
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")

#Reading the data  
current_directory=os.getcwd()
root_directory=os.path.abspath(os.path.join(current_directory, os.pardir)) #To get the parent directory of current directory
path=root_directory+"/Data"
data_path = os.path.join(path, 'globalterrorismdb_0616dist.csv')
```

```python
data = pd.read_csv(data_path,encoding = "ISO-8859-1")
```

```python
#creating dataframe for weapon type and its count
df  = data.groupby(['weaptype1_txt'])['eventid'].count().reset_index().sort_values('eventid', ascending = False)
df.columns = ['WeaponType','Attacks']
df.drop(df.index[[2,3]])
```

```python
#creating a plot for the weapon types and the major number of weapons used
colors = ['yellowgreen', 'gold', 'lightskyblue','red','blue','black','cyan','darkgreen','gray','white','purple']
patches, texts = plt.pie(df.Attacks, colors=colors, startangle=90)
plt.legend(patches,df.WeaponType, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
