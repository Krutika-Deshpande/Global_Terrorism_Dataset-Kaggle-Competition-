*************************
3rd Analysis - Successful Attacks based on Target
*************************

The terrorist attacks occured were successful and some also failed.
The targets attacked by the terrorist attacks fall under various common categories which include 
government organizations, educational institutions, airport/ aircraft ,etc
Those attacks which were successful based on the target types are plotted.
The plot depicts the attacks most succeeded on the target types.



```python
# importing libraries such as numpy,pandas and matplotlib
import os
import pandas as pd
from pandas import DataFrame
import datetime
import calendar
import warnings
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")
import mpl_toolkits.basemap
from mpl_toolkits.basemap import Basemap

current_directory=os.getcwd()
root_directory=os.path.abspath(os.path.join(current_directory, os.pardir)) #To get the parent directory of current directory
path=root_directory+"/Data"
data_path = os.path.join(path, 'globalterrorismdb_0616dist.csv')
```

```python
# Target Types and the successful attacks based on it
data = pd.read_csv(data_path,encoding = "ISO-8859-1")
df = DataFrame()
df = data.groupby('targtype1_txt')['success'].count().reset_index()
df
```

```python
# plot of successful events and target types
g = df.plot(kind='bar', color='green', alpha=0.7, align = 'center')
g.set_xticklabels(df['targtype1_txt'], rotation=90)
g.set_ylabel("Percentage of Success based on Target Type")
g.set_xlabel("Target Type")
g.set_title("Successful attacks on specific targets")
plt.show()
```
