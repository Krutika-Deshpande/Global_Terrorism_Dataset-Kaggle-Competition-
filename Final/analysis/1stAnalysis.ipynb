The First Analysis finds country with the maximum number of attacks occurred around the globe.
Iraq is the country with the maximum number of terrorist attacks with the count 18770 for the period of 1970-2015.
Terrorist attacks happened were of various types in Iraq and also worldwide.
Number of people injured and killed are given in the dataset.
This analysis is about the number of people injured and killed based on the attack types in Iraq.

```python

# Importing the necessary libaries like pandas, numpy , matplotlib 
import os
import pandas as pd
from pandas import DataFrame
import datetime
import calendar
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")

# Reading the dataset through the folder 'Data'
current_directory=os.getcwd()
root_directory=os.path.abspath(os.path.join(current_directory, os.pardir)) #To get the parent directory of current directory
path=root_directory+"/Data"
data_path = os.path.join(path, 'globalterrorismdb_0616dist.csv')
```

```python
# Reading the csv and fetching required columns.
data = pd.read_csv(data_path,encoding = "ISO-8859-1",usecols = [0, 1, 2, 3, 8, 11, 13, 14, 35, 84, 100, 103])
#Number of Attacks in every country around the globe.
Max_Attack_Country = data.groupby(['country_txt'])['eventid'].count().reset_index().sort_values('eventid', ascending = False)
Max_Attack_Country.columns = ['Country','Number_Of_Attacks']
#Country with the maximum number of terrorist Attacks
print('The country with maximum number of terrorist attacks is ' + (Max_Attack_Country['Country'].iloc[0]) +' '+ 'with ' + str(Max_Attack_Country['Number_Of_Attacks'].iloc[0]) + ' ' + 'attacks.')
```

```python
# Creating a dataframe with required columns
df1 = pd.read_csv(data_path,encoding = "ISO-8859-1",usecols = [0, 1, 2, 3, 8,29, 100, 103])
df1 = df1.rename(columns = {'eventid':'Event_ID','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','attacktype1_txt':'Attack_Type', 'nkill':'Fatalities', 'nwound':'Injured'})
df1.head()
```

```python
#Dataframe of Country Iraq 
#Day with value = 0 is set to 1
#Date is calculated with the year, month and day columns and a new date column is created.

Iraq_Attack = df1[df1.Country == 'Iraq']
Iraq_Attack['Day'][Iraq_Attack.Day == 0] = 1
Iraq_Attack['Date'] = pd.to_datetime(Iraq_Attack[['Year','Month','Day']])
Iraq_Attack.drop(['Year','Month','Day'], axis = 1, inplace = True)
Iraq_Attack_Columns = Iraq_Attack.columns.tolist() 
Iraq_Attack_Columns.insert(1, Iraq_Attack_Columns.pop(Iraq_Attack_Columns.index('Date')))
Iraq_Attack = Iraq_Attack.reindex(columns = Iraq_Attack_Columns)
```

```python
df = DataFrame()
df= Iraq_Attack.groupby('Attack_Type')['Attack_Type'].count().reset_index(name = "Attack")
#df['Fatalities'] = np.asarray(Iraq_Attack.groupby('Attack_Type')['Fatalities'].count())
f= np.asarray(Iraq_Attack.groupby('Attack_Type')['Fatalities'].count())
df['Fatalities'] = (f/f.sum()) *100;
i= np.asarray(Iraq_Attack.groupby('Attack_Type')['Injured'].count())
df['Injured'] = (i/i.sum())*100;
del df['Attack']
df #Percentage of killed and injured is calculated for the attack types
```

```python
# Plot for the percentage of people killed and injured in Iraq based on the Attack Types 
colors = ('red', 'blue')
g = df.plot(kind='bar', color=colors, alpha=0.7, align = 'center')
g.set_xticklabels(df['Attack_Type'], rotation=90)
#g.set_xticks()
#g.despine(left=True)
g.set_ylabel("Percentage of People killed and Injured in Iraq")
g.set_xlabel("Attack Types in Iraq")
g.set_title("People killed and injured in Iraq based on the attack type")
plt.show()
fig =plt.figure()
fig.savefig('1st.png') 
```
