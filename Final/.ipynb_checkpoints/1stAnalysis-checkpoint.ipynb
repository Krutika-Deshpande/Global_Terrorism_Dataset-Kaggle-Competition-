```python
df1 = pd.read_csv('\\Users\\kruts\\Documents\\GitHub\\deshpande_krutika_001280718_Spring2017\\Final\\data\\globalterrorismdb_0616dist.csv',encoding = "ISO-8859-1",usecols = [0, 1, 2, 3, 8, 11, 13, 14, 18,19,20,21,26,27,29,35, 84, 100, 103])
df1 = df1.rename(columns = {'eventid':'Event_ID','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','latitude':'Latitude','longitude':'Longitude','summary':'Attack_Description','crit1':'Criteria_1','crit2':'Criteria_2','crit3':'Criteria_3','attacktype1_txt':'Attack_Type','targtype1_txt':'Target','weaptype1_txt':'Weapon_Type', 'nkill':'Fatalities', 'nwound':'Injured'})
```

```python
#First Analysis 
Iraq_Attack = df1[df1.Country == 'Iraq']
Iraq_Attack['Day'][Iraq_Attack.Day == 0] = 1
Iraq_Attack['Date'] = pd.to_datetime(Iraq_Attack[['Year','Month','Day']])
Iraq_Attack.drop(['Year','Month','Day'], axis = 1, inplace = True)
Iraq_Attack_Columns = Iraq_Attack.columns.tolist() 
Iraq_Attack_Columns.insert(1, Iraq_Attack_Columns.pop(Iraq_Attack_Columns.index('Date')))
Iraq_Attack = Iraq_Attack.reindex(columns = Iraq_Attack_Columns)
#Iraq_Attack = Iraq_Attack.groupby(['Attack_Type'])['Fatalities'].count()
sum(pd.isnull(Iraq_Attack['Attack_Type']))
Iraq_Attack.head()
```

```python

df = DataFrame()
df12 = DataFrame()
df= Iraq_Attack.groupby('Attack_Type')['Attack_Type'].count().reset_index(name = "Attack")
#df['Fatalities'] = np.asarray(Iraq_Attack.groupby('Attack_Type')['Fatalities'].count())
f= np.asarray(Iraq_Attack.groupby('Attack_Type')['Fatalities'].count())
df['Fatalities'] = (f/f.sum()) *100;
i= np.asarray(Iraq_Attack.groupby('Attack_Type')['Injured'].count())
df['Injured'] = (i/i.sum())*100;
del df['Attack']
df
```

```python
#data = pd.DataFrame(df, index=list(df['Attack_Type']))
colors = ('red', 'blue')
g = df.plot(kind='bar', color=colors, alpha=0.7, align = 'center')
g.set_xticklabels(df['Attack_Type'], rotation=90)
#g.set_xticks()
#g.despine(left=True)
g.set_ylabel("Percentage of People Died and Injured in Iraq")
g.set_xlabel("Attack Types in Iraq")
g.set_title("People died and injured in Iraq based on the attack type")
plt.show()
```
